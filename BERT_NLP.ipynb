 1. Preprocessing  Turning messy human text into BERT-friendly input

BERT doesnt understand normal text like hello bro whats up?.
It only understands numbers (token IDs) in a very specific format.

So preprocessing does things like:

 Tokenization

Splits words into subwords using WordPiece:
playing  play + ##ing

The ## prefix means:

This piece continues from the previous token, dont treat it as a new word.

 Add special tokens

BERT wants:

[CLS] at the start

[SEP] at the end

Or multiple [SEP]s for sentence pairs

Example:

[CLS] i love coding [SEP]

 Padding / Truncation

BERT needs fixed sequence length (like 128 tokens).

So preprocessing trims or pads your sentence.

 Convert tokens  IDs

BERT has a vocabulary file with ~30k tokens.

Each subword gets mapped to a unique number.

 Create attention masks

Tells BERT which tokens are real vs padding.

So preprocessing = make the text digestible for BERT.

 2. Encoder  The brain of BERT that understands language

Once the text is converted into numbers, the ENCODER (Transformer encoder stack) does all the heavy lifting:

 Self-Attention

It lets every word look at every other word to understand meaning.

bank in

I went to the bank to deposit money
vs
I sat on the bank of the river

BERT uses self-attention to figure out the correct meaning based on context.

 Multi-Head Attention

Multiple attention heads learn different types of relationships:

Syntax (subjectverb)

Long-range dependencies

Meaning

Position patterns

 Positional Encodings

Transformers dont know the order of words.
Positional embedding tells BERT the order.

 Deep layers learn deep meaning

BERT-Base = 12 encoder layers

BERT-Large = 24 encoder layers

Each layer extracts deeper semantic meaning.

 Putting it together
Stage	Why it exists	What it does
Preprocessing	BERT cant read raw text	Turns text  tokens  token IDs + masks
Encoder	This is the actual BERT model	Understands meaning, context, grammar, semantics
