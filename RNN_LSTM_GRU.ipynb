Traditional RNN :-
In an RNN, inputs are processed sequentially, meaning every word is strongly related to the previous ones in the sentence. However, when many words are given as input, the network often forgets the memory of the earlier words because during backpropagation the gradients shrink at each step, and as a result the weights corresponding to earlier inputs are not effectively updated. This is called vanishing gradient problem / short-term memory loss.


Output of one word is the input mixed with next new word ..and so on for the Sequence Learning of Human drivensentence.




Different Types of RNN :- NLP (Natural Language Processing)


Single RNN at different Time Stamps for language Processing :-
this is RNN network with a single layer at different time stamps -> output of one word is join/mixed with the input of next word ...becoz each word is related to eachotherinRNN


To remove the Vanishing Gradient Problem in RNN , we have 2 solutions :-  GRU, LSTM


In LSTM (Long Short-Term Memory) -> output = combination of ( long term (memorize special/main keywords) + short term (memorize only recent /current words) )memory line.



In GRU (Gated Recurrent Unit) -> single line for long and short termmemory .
